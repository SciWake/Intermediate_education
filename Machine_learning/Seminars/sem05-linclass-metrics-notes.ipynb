<h1 style="color:#008B8B">1. AUC-ROC</h1>

На лекции мы познакомились с такой важной метрикой качества бинарной классификации, как площадь под ROC-кривой (AUC-ROC). Напомним её определение. Рассмотрим задачу бинарной классификации с метками классов $\mathbb{Y} = \{ -1, +1\},$ и пусть задан некоторый алгоритм $b(x)$, позволяющий вычислять оценку принадлежности объекта $x$ положительному классу. AUC-ROC позволяет оценивать качество классификации для семейства алгоритмов следующего вида:
$$a(x; t) = \begin{cases}
		-1, \, b(x) \le  t,\\
		+1, \, b(x) > t,
	\end{cases}$$
    
т.е. алгоритмов, присваивающих метки объектам в соответствии с оценками $b(x)$, отсекая их по некоторому порогу $t$. Каждый алгоритм (получающийся при фиксации значения порога $t$) представляется точкой на плоскости (FPR, TPR), где

$$\text{FPR} = \frac{\text{FP}}{\text{FP}+\text{TN}} = \frac{\text{FP}}{\ell_-},$$

$$\text{TPR} = \frac{\text{TP}}{\text{TP}+\text{FN}} = \frac{\text{TP}}{\ell_+},$$
    
$\ell_-, \ell_+$ — количество объектов отрицательного и положительного классов соответственно. AUC-ROC, в свою очередь, является площадью под получившейся кривой.

Изучим подробнее некоторые важные свойства данной метрики. Критерий AUC-ROC имеет большое число интерпретаций — например, он равен вероятности того, что случайно выбранный положительный объект окажется позже случайно выбранного отрицательного объекта в ранжированном списке, порожденном $b(x)$. Разберем подробнее немного другую формулировку.

### Задача 1.1

В ранжировании часто используется функционал "доля дефектных пар". Его можно определить и для задачи бинарной классификации.

Пусть дан классификатор $b(x)$, который возвращает оценки принадлежности объектов классу +1, и пусть все значения $b(x_i), \, i= \overline{1, \ell},$ для некоторой выборки $X = \{\left(x_i, y_i\right)\}_{i=1}^\ell$ различны. 

Отсортируем все объекты по возрастанию ответа классификатора: $b(x_{(1)}) < \dots < b(x_{(\ell)})$. Где индекс $(1)$ означает объект в упорядоченной выборке, который расположен на первом месте. Обозначим истинные ответы на этих объектах через $y_{(1)}, \dots, y_{(\ell)}.$ 

Тогда доля дефектных пар записывается как
	
$$\text{DP}(b, X) = \frac{2}{\ell(\ell-1)} \sum_{i < j}^\ell [y_{(i)} > y_{(j)}].$$

Как данный функционал связан с AUC-ROC?

<h4 align='center'>Решение</h4>
    
Всего возможных порогов получится $\ell + 1$. Если все пороги изобразить на числовой прямой, тогда в начале у нас будет находится объект $b(x_{(1)})$ в котором модель меньше всего уверна, что он относится к положительному классу.

Для начала разберем процедуру построения ROC-кривой. Сперва все объекты сортируются по неубыванию оценки $b(x)$, тем самым формируя список $x_{(1)}, \dots, x_{(\ell)}.$ Заметим, что для построения ROC-кривой достаточно рассмотреть $(\ell+1)$ различных значений порога $t$, соответствующих всем различным способам классификации выборки, порожденным алгоритмом $b(x)$, - например, в качестве таких порогов можно рассмотреть следующий набор:

$\large t_\ell = b(x_{(\ell)}) + 1$ - все объекты которые ниже данного порога, будут относится к отрицательным. Значит все объект будут отрицательными и ни один объект не будет отнесен к положительному.

$\large t_i = \frac{b(x_{(i)}) + b(x_{(i+1)})}{2} , \, i = \overline{1, \ell -1}$

$\large t_0 = b(x_{(1)}) - 1.$ -  в этом случае все объекты выше данного порога будут отнесены к положительным.

Зафиксируем значение порога $t = t_\ell =  b(x_{(\ell)}) + 1$, в этом случае имеем

$$\text{FPR} = \frac{\text{FP}}{\ell_-} = \frac{0}{\ell_-} = 0,$$

$$\text{TPR} = \frac{\text{TP}}{\ell_+} = \frac{0}{\ell_+} = 0.$$

Таким образом, алгоритму $a(x; t_\ell)$ соответствует точка (0; 0) на плоскости, откуда начинается построение ROC-кривой.

Будем перебирать пороги в порядке невозрастания их значения, начиная с $t_{\ell}$. Пусть мы хотим уменьшить значение порога с $t_i$ до $t_{i-1}$. При этом классификация объекта $x_{(i)}$ (и только его) изменится с отрицательной на положительную. Рассмотрим 2 случая.

<img src="img/5_1.png">

* $y_{(i)} = +1.$ В этом случае классификатор начнет верно классифицировать объект, на котором ранее допускал ошибку, при этом FPR не изменится, а TPR повысится на $\frac{1}{\ell_+}.$

* $y_{(i)} = -1.$ В этом случае классификатор начнет ошибаться на объекте, который ранее классифицировал верно, при этом TPR не изменится, а FPR повысится на  $\frac{1}{\ell_-}.$

<img src="img/5_2.png" width="500" height="500">

**Площадь одного бина**

Теперь рассмотрим, как при этом изменяется AUC-ROC. Заметим, что область под ROC-кривой состоит из непересекающихся прямоугольников, каждый из которых снизу ограничен осью FPR, а сверху - одним из горизонтальных отрезков, соответствующих второму из рассмотренных случаев. Поэтому каждый раз, когда имеет место второй случай, к текущей накопленной площади под кривой (которая изначально в точке (0; 0) равна 0) добавляется площадь прямоугольника, горизонтальные стороны которого равны 

$\large \frac{1}{\ell_-}$

Для вертикальной стороны необходимо посчитать то, сколько раз до этого мы сдвигались на $\frac{1}{\ell_+}$. Так как мы двигаем по оси $\text{TPR}$ только тогда, когда класс был отнесен к положительному правильно, значит можем просумировать все предыдущие ответы $y$ (то есть справа) по $j$:

$\large \sum\limits_{j=i+1}^\ell \frac{1}{\ell_+} [y_{(j)} = +1] = \frac{1}{\ell_+} \sum\limits_{j=i+1}^{\ell} [y_{(j)} = +1]$ - вертикальная сторона бина (доля уже рассмотренных положительных объектов среди всех положительных).

Суммируем все истенные ответы по $j$, где $j = i + 1$ или $i < j$. Так как мы упорядочили объекты по возрастанию прогноза модели, а пороги мы сдвигали как $t_{i}, t_{i- 1}\ldots$, тогда для рассчета всех предыдущих объектов, мы должны идти в обратную сторону $j = i+1$ по объектам. Движимся мы по объектам так как они отсортированы по выходам модели и это то же самое что идти по порогам.

Добавим сюда горизонтальную сторону и в этом случае текущее значение AUC-ROC увеличивается на:

$\large \text{bin} =\frac{1}{\ell_- \ell_+} \sum\limits_{j=i+1}^\ell [y_{(j)} = +1].$

**Посчитаем по всем бинам**

Новый прямоугольник появляется тогда, когда мы встречаем отрицательный $y$, значит, добавим сумму по всем $i$, где $y_{(i)}$ будет принимать минус еденицу и для этого $i$ считаем площадь бина $bin$, формулу которой мы вывели выше.

Итого, финальное значение AUC-ROC можно посчитать следующим образом:

$\large \text{AUC} = \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i = 1}^{\ell} [y_{(i)} = -1] \sum\limits_{j = i + 1}^{\ell} [y_{(j)} = +1] =$

Оба индикатора под знаком суммы принимают 1 или 0. Вклад в сумму вносится тогда, когда индикатор равен 1. Так как первый индикатор будет равен 1 тогда, когда $y_{(i)}$ будет принимать значение 0. А воторой индикатор будет равен 1 тогда, когда $y_{(j)}$ будет равен 1. А это можно перезаписать в следующем виде $y_{(i)} < y_{(j)}$:

$\large = \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i = 1}^{\ell} \sum\limits_{j = i + 1}^{\ell} [y_{(i)} < y_{(j)}] =$

Можем схлопнуть две суммы. Дальше от еденицы отнимаем противоположный исход (два противоположных условия, а все три комбинации в сумме дадут 1. Значит 1 минус два условия равно $[y_{(i)} < y_{(j)}]$):

$\large = \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} (1 - [y_{(i)} = y_{(j)}] - [y_{(i)} > y_{(j)}]) =$

Дальше вынесим каждый индикатор под сумму:

$\large = \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} (1 - [y_{(i)} = y_{(j)}]) - \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} [y_{(i)} > y_{(j)}] =$

Мы знаем что $(1 - [y_{(i)} = y_{(j)}]) = [y_{(i)} \ne y_{(j)}]$:

$\large = \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} ([y_{(i)} \ne y_{(j)}]) - \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} [y_{(i)} > y_{(j)}] =$

Раскроем эту сумму $\sum\limits_{i < j}^{\ell} ([y_{(i)} \ne y_{(j)}])$. Для выполнения условия $[y_{(i)} \ne y_{(j)}]$ необходимо чтобы один из классов был положителен другой отрицателен. Значит для каждого $y_{(i)}$ условие будет удовлетворятся столько раз, сколько у нас объектов в противоположном классе для $y_{(j)}$. Значит мы добавим столько индикаторов, сколько у нас объектов другого класса начиная с $i$, а это можно записать как $\ell_{+} \cdot \ell_{-}$:

$\large = \frac{\ell_{+} \ell_{-}}{\ell_{+} \ell_{-}} - \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} [y_{(i)} > y_{(j)}] = 1 - \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} [y_{(i)} > y_{(j)}].$

Тем самым, AUC-ROC отражает долю пар, которые относительно друг друга ранжированы неправильно. Отсюда получаем, что AUC-ROC и доля дефектных пар связаны следующим соотношением:
$$\text{DP}(b, X) = \frac{2 \ell_- \ell_+}{\ell (\ell -1)} (1 - \text{AUC} (b, X))).$$

Значит, чем больше дефектных пар, тем больше будет AUC-ROC и наоборот.

### Пояснение формулы

$\large \text{AUC} = 1 - \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} [y_{(i)} > y_{(j)}]$

Здесь все элементы упорядочены по предсказанию классификатора. И в сумме, у нас все элементы, такие, которые упорядочены неправильно. $i < j$ означает, что $b(x_{i}) < b(x_{j})$, но приэтом $y_{(i)} > y_{(j)}$, то есть два объекта относительно друг друга стоят неправильно.

<img src="img/5_3.svg">

Если мы берем порог 

* $t_{i + 1}$, тогда $b(x_{i}), b(x_{j})$ оба отнесены к отрицательному классу, что плохо для нас.

* $t_{i}$, тогда $b(x_{j})$ будет отнесен к положительному классу, а $b(x_{i})$ к отрицательному, что неправильно для двух объектов.

* $t_{i - 1}$, тогда $b(x_{i}), b(x_{j})$ оба объекта будут отнесены к положительному классу, что плохо для нас.

И можем сделать выводы, что здесь нет идеального порога, так как два объекта стоят неправильно относительно друг друга. Но если их поменять местами (+1 справа, а -1 слева), тогда мы смогли бы взять порог $t_{i}$.



### Когда выходы модели совпадают

Заметим, что в случае, когда несколько объектов выборки имеют равные значения $b(x)$, при уменьшении значения порога с $t_i > b(x)$  до $t_{i-1} < b(x),$ где $x$ — один из таких объектов, изменение значений FPR и TPR происходит одновременно, поэтому  соответствующий участок ROC-кривой будет наклонным, а не горизонтальным или вертикальным.

При переходе от порога $t_i > b(x)$  до $t_{i-1} < b(x)$, как изменится $\text{TP}$ и $\text{FP}$? На их изменение будет влияет значение того, чему будет равняется истенный ответ $y$. И если у нас два объекта находятся в точке $x$, где в этой точке мы имеем $b(x_{(i)}) == b(x_{(j)})$ могут возникать следующие ситуации, которые будут оказывать воздействие на площадь:

* $y_{(i)} = +1 \quad y_{(j)} = +1$ тогда $\text{TP}_{i-1} = \text{TP}_{i} + \frac{2}{\ell_{+}}$

* $y_{(i)} = -1 \quad y_{(j)} = -1$ тогда $\text{FP}_{i-1} = \text{FP}_{i} + \frac{2}{\ell_{-}}$

* $y_{(i)} = +1 \quad y_{(j)} = -1$ или $y_{(i)} = -1 \quad y_{(j)} = +1$ тогда 
    
    $\text{TP}_{i-1} = \text{TP}_{i} + \frac{1}{\ell_{+}}$
    
    $\text{FP}_{i-1} = \text{FP}_{i} + \frac{1}{\ell_{-}}$

### Задача 1.3

Пусть $b(x)$ — классификатор, предсказывающий оценку принадлежности объекта $x$ классу +1 таким образом, что для некоторой выборки $X$ он равновероятно выдаёт на её объектах одну из всех возможных перестановок. Чему равно матожидание AUC-ROC этого классификатора?

<h4 align='center'>Решение</h4>

Как было показано в задаче 1.1, для AUC-ROC верно

$\text{AUC} = \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} [y_{(i)} = -1] [y_{(j)} = +1],$ поэтому

$\mathbb{E}\text{AUC} = \frac{1}{\ell_{+} \ell_{-}} \sum\limits_{i < j}^{\ell} \mathbb{E}([y_{(i)} = -1] [y_{(j)} = +1]).$

Заметим, что величина $[y_{(i)} = -1] [y_{(j)} = +1]$ принимает значения 0 и 1, поэтому математическое ожидание индикатора равно вероятности $\mathbb{E}([y_{(i)} = -1] [y_{(j)} = +1]) = \mathbb{P}(y_{(i)} = -1 , \, y_{(j)} = +1) = \frac{\ell_- \ell_+ (\ell-2)!}{\ell!} = \frac{\ell_- \ell_+}{\ell (\ell-1)}.$

Как мы получили $\frac{\ell_- \ell_+ (\ell-2)!}{\ell!}$? Мы знаем что среди всех элементов есть пара $y_{(i)} = -1, y_{(j)} = +1$ тогда на место $y_{(i)}$ можем расположить любой отрицательный объект, которых $\ell_{-}$, а для $y_{(j)}$ можем расположить любой положительный объект, которых $\ell_{+}$, ну а на остальные позиции можем расположить $\ell - 2$ элемтов, а это можно сделать $(\ell - 2)!$ способами. И делим это на общее число перестановок $\ell!$.

Отсюда имеем

$\mathbb{E}AUC = \frac{1}{\ell_- \ell_+} \sum\limits_{i<j}^\ell \frac{\ell_- \ell_+}{\ell (\ell-1)} = 
	\frac{\ell (\ell -1)}{2} \frac{1}{\ell (\ell-1)} = \frac{1}{2}.$

Итого, можем заметить, что значение AUC-ROC, близкое к $\frac{1}{2},$ означает, что классификатор близок к случайному, тогда как значение, равное 1, означает, что классификатор безошибочно классифицирует объекты при некотором значении порога.
